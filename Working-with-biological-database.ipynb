```python
import Bio
from Bio import Seq
from Bio import SeqIO
```

```python
x = SeqIO.parse('y.fasta', 'fasta')
print(x)

for record in x:
    print(record.id)
    print(record.description)
    print(record.seq)

x = SeqIO.read('x.fasta', 'fasta')
print(x.seq)
```

```python
from Bio.SeqUtils import gc_fraction as GC
from Bio.SeqUtils import MeltingTemp as MT
from matplotlib import pyplot as plt

seqX = x.seq
seqA = seqX[:round(len(seqX)/2)]
seqB = seqX[round(len(seqX)/2):]

gcA = GC(seqA)
gcB = GC(seqB)
mtA = MT.Tm_NN(seqA)
mtB = MT.Tm_NN(seqB)

fig, axs = plt.subplots(1, 2)
axs[0].bar(['GC A', 'GC B'], [gcA, gcB])
axs[1].bar(['MT A', 'MT B'], [mtA, mtB])
plt.show()
```
